```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def set_axes(ax):
    xmin, xmax= ax.get_xlim()
    ymin, ymax= ax.get_ylim()
    
    xticks = np.arange(int(xmin), int(xmax)+1)
    yticks = np.arange(int(ymin), int(ymax)+1)

    for xt in xticks:
        ax.axvline(x=xt, ymin=ymin, ymax=ymax, lw=1, ls='-.' if xt != 0 else '-', color='k'  if xt == 0 else 'lightgrey')

    for yt in yticks:
        ax.axhline(y=yt, lw=1, ls='-.' if yt != 0 else '-', color='k' if yt == 0 else 'lightgrey')


    ax.set_xticks(ticks=xticks)
    ax.set_yticks(ticks=yticks)
```

```python
def add_vector(ax, coord, label, color='k'):
        (x,y), (dx,dy) = coord
        ax.arrow(x, y, dx, dy, lw=3, head_width=0.1, head_length=0.2, length_includes_head=True, fc=color, ec=color)
        ax.annotate(label, xy=[x+dx/2, y+dy/2+0.3], size=30, color=color)
    
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.gca()
ax.set_xlim(-2, 18)
ax.set_ylim(-2, 8)
add_vector(ax, [[0, 0], [2, 1]], r'$\vec{a}$', color='k')
add_vector(ax, [[3, 3], [1, 1]], r'$\vec{b}$', color='k')
add_vector(ax, [[12, 4], [2, 1]], r'$\vec{c}$', color='k')
add_vector(ax, [[5, 7], [1, 1]], r'$\vec{d}$', color='r')
add_vector(ax, [[3, 6], [-2, 2]], r'$\vec{e}$', color='r')
add_vector(ax, [[-1, 5], [1, 2]], r'$\vec{f}$', color='r')
add_vector(ax, [[14, 6], [-2, 2]], r'$\vec{g}$', color='g')
add_vector(ax, [[-1, 3], [2, 0]], r'$\vec{h}$', color='g')
add_vector(ax, [[6, 5], [2, -1]], r'$\vec{i}$', color='k')
add_vector(ax, [[7, 3], [1, -1]], r'$\vec{j}$', color='k')
add_vector(ax, [[8, 2], [2, 0]], r'$\vec{k}$', color='k')
add_vector(ax, [[8, 7], [1, -2]], r'$\vec{l}$', color='r')
add_vector(ax, [[10, 5], [1, -2]], r'$\vec{m}$', color='r')
add_vector(ax, [[11, 2], [1,-2]], r'$\vec{n}$', color='r')
add_vector(ax, [[15, 4], [1, -2]], r'$\vec{r}$', color='g')
add_vector(ax, [[13, 0], [-1, 2]], r'$\vec{p}$', color='g')

set_axes(ax)
plt.savefig('pics/vector_equal1.png')
plt.show()
```

Znajdź jednakowe wektory <br>
 $1. \quad\vec{n} = \vec{p}\quad\quad   2.\quad \vec{a} = \vec{c}\quad \quad3.\quad \vec{h} = \vec{k}$ <br>
 $4. \quad \vec{j} = \vec{b}\quad\quad  5. \quad \vec{a} = \vec{h}\quad \quad 6.\quad \vec{k} = \vec{h}$ <br>
 $7. \quad \vec{b} = \vec{d}\quad \quad  8. \quad \vec{a} = \vec{h}\quad \quad 9. \quad\vec{l} = \vec{n}= \vec{r}$ <br>   

```python
fig = plt.figure(figsize=(5,5))
ax = fig.gca()
ax.set_xlim(-1, 4)
ax.set_ylim(-1, 4)
add_vector(ax, [[-1, 1], [3, 2]], r'$\vec{a}$', color='r')
add_vector(ax, [[2, 3], [1, -3]], r'$\vec{b}$', color='g')
add_vector(ax, [[-1, 1], [4, -1]], r'$\vec{c}$', color='b')

set_axes(ax)
plt.savefig('pics/vector_sum1.png')
plt.show()
```
