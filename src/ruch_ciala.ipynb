```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def set_axes(ax):
    xmin, xmax= ax.get_xlim()
    ymin, ymax= ax.get_ylim()
    
    xticks = np.arange(int(xmin), int(xmax)+1)
    yticks = np.arange(int(ymin), int(ymax)+1)

    for xt in xticks:
        ax.axvline(x=xt, ymin=ymin, ymax=ymax, lw=1, ls='-.' if xt != 0 else '-', color='k'  if xt == 0 else 'lightgrey')

    for yt in yticks:
        ax.axhline(y=yt, lw=1, ls='-.' if yt != 0 else '-', color='k' if yt == 0 else 'lightgrey')


    ax.set_xticks(ticks=xticks)
    ax.set_yticks(ticks=yticks)
```

```python
def add_vector(ax, coord, label, color='k'):
        (x,y), (dx,dy) = coord
        ax.arrow(x, y, dx, dy, lw=3, head_width=0.1, head_length=0.2, length_includes_head=True, fc=color, ec=color)
        ax.annotate(label, xy=[x+dx/2, y+dy/2+0.3], size=30, color=color)
    
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.gca()
ax.set_xlim(0, 10)
ax.set_ylim(0, 120)

xmin, xmax= ax.get_xlim()
ymin, ymax= ax.get_ylim()

xticks = np.arange(int(xmin), int(xmax)+1)
yticks = np.arange(int(ymin), int(ymax)+1, 10)

for xt in xticks:
    ax.axvline(x=xt, ymin=ymin, ymax=ymax, lw=1, ls='-.' if xt != 0 else '-', color='k'  if xt == 0 else 'lightgrey')

for yt in yticks:
    ax.axhline(y=yt, lw=1, ls='-.' if yt != 0 else '-', color='k' if yt == 0 else 'lightgrey')

ax.set_xticks(ticks=xticks)
ax.set_yticks(ticks=yticks)

ax.set_xlabel('Czas, h')
ax.set_ylabel('Droga, km')

plt.plot([0,2,3,6,7,8,10],[0,30,30,80,80,100,120], c='r')

plt.savefig('pics/ruch1.png')
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.gca()
ax.set_xlim(0, 5)
ax.set_ylim(0, 15)

xmin, xmax= ax.get_xlim()
ymin, ymax= ax.get_ylim()

xticks = np.arange(int(xmin), int(xmax)+1)
yticks = np.arange(int(ymin), int(ymax)+1)

for xt in xticks:
    ax.axvline(x=xt, ymin=ymin, ymax=ymax, lw=1, ls='-.' if xt != 0 else '-', color='k'  if xt == 0 else 'lightgrey')

for yt in yticks:
    ax.axhline(y=yt, lw=1, ls='-.' if yt != 0 else '-', color='k' if yt == 0 else 'lightgrey')

ax.set_xticks(ticks=xticks)
ax.set_yticks(ticks=yticks)

ax.set_xlabel('Czas, h', fontsize=16)
ax.set_ylabel('Droga, km', fontsize=16)

t = np.linspace(0,10)
plt.plot(t, 2*t, c='r', label='Samochod A')
plt.plot(t, 0.5*t**2, c='g', label='Samochod B')

plt.legend()
plt.savefig('pics/ruch8.png')
plt.show()
```
