```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
plt.xticks(ticks=[1,2], labels=['V', '2V'])
plt.yticks(ticks=[1,2], labels=[r'$\frac{p}{2}$','p'])

plt.plot([1,2], [2,1], c='k', lw=2)
plt.xlabel('objętność, V')
plt.ylabel('cieśnienie, p')

plt.plot([1,1], [0,2], c='k', lw=1, ls='--')
plt.plot([2,2], [0,1], c='k', lw=1, ls='--')
plt.plot([0,1], [2,2], c='k', lw=1, ls='--')
plt.plot([0,2], [1,1], c='k', lw=1, ls='--')

plt.xlim([0,2.5])
plt.ylim([0,2.5])

plt.text(1.05,2.05,'1')
plt.text(2.05,1.05,'2')
plt.savefig('pics/term_1.png')
plt.show()
```
