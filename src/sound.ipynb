```python
volume = 0.45     # range [0.0, 1.0]
fs = 44100       # sampling rate, Hz, must be integer
duration = 1   # in seconds, may be float
```

```python
c4 = 261.63
c4_ = 277.18
d4 = 293.66
d4_= 311.13
e4 = 329.63
f4 = 349.23
f4_ = 369.99
g4 = 392.00
g4_ = 415.30
a4 = 440.00
a4_ = 466.16
b4 = 493.88
c5 = 523.25
```

```python
C4 = (np.sin(2*np.pi*np.arange(fs*duration)*c4/fs)).astype(np.float32)
C4_ = (np.sin(2*np.pi*np.arange(fs*duration)*c4_/fs)).astype(np.float32)
D4 = (np.sin(2*np.pi*np.arange(fs*duration)*d4/fs)).astype(np.float32)
D4_= (np.sin(2*np.pi*np.arange(fs*duration)*d4_/fs)).astype(np.float32)
E4 = (np.sin(2*np.pi*np.arange(fs*duration)*e4/fs)).astype(np.float32)
F4 = (np.sin(2*np.pi*np.arange(fs*duration)*f4/fs)).astype(np.float32)
F4_ = (np.sin(2*np.pi*np.arange(fs*duration)*f4_/fs)).astype(np.float32)
G4 = (np.sin(2*np.pi*np.arange(fs*duration)*g4/fs)).astype(np.float32)
G4_ = (np.sin(2*np.pi*np.arange(fs*duration)*g4_/fs)).astype(np.float32)
A4 = (np.sin(2*np.pi*np.arange(fs*duration)*a4/fs)).astype(np.float32)
A4_ = (np.sin(2*np.pi*np.arange(fs*duration)*a4_/fs)).astype(np.float32)
B4 = (np.sin(2*np.pi*np.arange(fs*duration)*b4/fs)).astype(np.float32)
C5 = (np.sin(2*np.pi*np.arange(fs*duration)*c5/fs)).astype(np.float32)
```

```python
gamma = np.concatenate([C4, C4_, D4, D4_, E4, F4, F4_, G4, G4_, A4, A4_, B4])
```

```python
import numpy as np
from matplotlib import pyplot as plt

SAMPLE_RATE = 44100  # Hertz
DURATION = 5  # Seconds

def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq
    # 2pi because np.sin takes radians
    y = np.sin((2 * np.pi) * frequencies)
    return x, y

# Generate a 2 hertz sine wave that lasts for 5 seconds
x, y = generate_sine_wave(2, SAMPLE_RATE, DURATION)
plt.plot(x, y)
plt.show()
```
